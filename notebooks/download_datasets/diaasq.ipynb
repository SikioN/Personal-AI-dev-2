```python
import numpy as np
from typing import List, Dict
import json
import os
```

```python
SAVE_STATS_PATH = '../../data/qa_datasets/diaasq/stats.json'
QAPAIRS_DIR_PATH = '../../data/qa_datasets/diaasq/qa_eval'
CONTEXTS_PATH = '../../data/qa_datasets/diaasq/Augment_DiaASQ.json'
```

```python
def compute_stats(numbers: List[int]) -> Dict[str, float]:
    return {'median': round(np.median(numbers),5),
            'mean': round(np.mean(numbers),5),
            'std': round(np.std(numbers),5)}
```

```python
with open(CONTEXTS_PATH, 'r', encoding='utf-8') as fd:
    contexts = list(map(lambda item: item['text_dialog'], json.loads(fd.read())['data']))
```

```python
qaclasss_files = os.listdir(QAPAIRS_DIR_PATH)
questions, answers = [], []
for qafile in qaclasss_files:
    qafile_path = f"{QAPAIRS_DIR_PATH}/{qafile}"
    with open(qafile_path, 'r', encoding='utf-8') as fd:
        qa_pairs = json.loads(fd.read())
    for pair in qa_pairs:
        questions.append(pair['question'])
        answers.append(pair['answer'])
```

```python
filtered_ds_statistics = {
    'contexts_amount': len(contexts),
    'contexts_len': compute_stats(list(map(lambda v: len(v), contexts))),
    'qa_pairs_amount': len(questions),
    'answers_len':  compute_stats(list(map(lambda v: len(v), answers))),
    'questions_len': compute_stats(list(map(lambda v: len(v), questions)))
}
```

```python
with open(SAVE_STATS_PATH, 'w', encoding='utf-8') as fd:
    json.dump(filtered_ds_statistics, fd, indent=1, ensure_ascii=False)
```
